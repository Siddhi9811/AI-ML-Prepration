```python
import pandas as pd

# Example Dataset
data = {
    'Name': ['Riya', 'Amit', 'Karan', 'Siddhi', 'Neha'],
    'Age': [20, 22, 19, None, 21],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Female'],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Pune', None],
    'Marks': [88, 75, 90, 95, 80]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
print(df.isnull().sum())
```

```python
# Fill missing Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing City with most frequent value (mode)
df['City'].fillna(df['City'].mode()[0], inplace=True)

```

```python
print(df)
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['City'] = le.fit_transform(df['City'])
print(df)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Age', 'Marks']] = scaler.fit_transform(df[['Age', 'Marks']])
print(df)
```

```python
from sklearn.model_selection import train_test_split

X = df[['Age', 'Gender', 'City', 'Marks']]
y = [1, 0, 1, 1, 0]  # Example labels (1 = Pass, 0 = Fail)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training Data:\n", X_train)
print("Testing Data:\n", X_test)
```
