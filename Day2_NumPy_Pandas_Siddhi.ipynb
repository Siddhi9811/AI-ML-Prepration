```python
import numpy as np
arr = np.random.randint(10, 100, 10)
print("Array:", arr)
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
reshaped = arr.reshape(2,5)
print("Reshaped (2x5):\n", reshaped)
print("Second Row:", reshaped[1])
print("Identity Matrix:\n", np.eye(3))
```

```python
import pandas as pd
data = {
    'Name': ['Amit', 'Riya', 'Karan', 'Siddhi', 'Anya'],
    'Marks': [88, 92, 79, 95, 84],
    'Age': [20, 21, 19, 22, 20],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Delhi', 'Pune']
}
df = pd.DataFrame(data)
print(df)
print(df.describe())
print(df[df['Marks'] > 85])
df['Grade'] = ['A' if m >= 90 else 'B' for m in df['Marks']]
print(df)
print("Average Marks:", df['Marks'].mean())
```
