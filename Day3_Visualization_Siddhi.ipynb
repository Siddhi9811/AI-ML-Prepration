```python
#LINE CHART
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y = [10, 12, 8, 14, 20]
plt.plot(x,y,marker='o',color='purple',linestyle='--')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt
categories = ['Apple', 'Banana', 'Mango', 'Grapes']
values = [40, 70, 30, 90]
plt.bar(categories,values,color='teal')
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Quantity Sold')
plt.show()

```

```python
# Horizontal Bar Chart
import matplotlib.pyplot as plt
categories = ['Apple', 'Banana', 'Mango', 'Grapes']
values = [40, 70, 30, 90]
plt.barh(categories,values,color='teal')
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Quantity Sold')
plt.show()

```

```python
#Pie Chart
import matplotlib.pyplot as plt
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 20, 30]

plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()

```

```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('tips')
sns.histplot(df['total_bill'], bins=10, kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill by Day')
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.title('Count of Days')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
plt.title('Tip vs Total Bill')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')

# Calculate the mean tip for each day
mean_tip_per_day = df.groupby('day')['tip'].mean()

# Plot using plt.bar
plt.bar(x=mean_tip_per_day.index, height=mean_tip_per_day.values, color='pink')
plt.title('Mean Tip by Day')
plt.xlabel('Day')
plt.ylabel('Mean Tip')
plt.show()
```

```python
plt.scatter(x='total_bill', y='tip', data=df)
plt.title('Tip vs Total Bill')
plt.show()
```

```python
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip by Gender')
plt.show()
```
