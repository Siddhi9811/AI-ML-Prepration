```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Dataset
data = {
    'Study_Hours': [1, 2, 3, 4, 5, 6, 7, 8],
    'Sleep_Hours': [9, 8, 7, 6, 6, 5, 4, 3],
    'Pass': [0, 0, 0, 1, 1, 1, 1, 1]  # 1=Pass, 0=Fail
}

df = pd.DataFrame(data)
print(df)
```

```python
X = df[['Study_Hours', 'Sleep_Hours']]
y = df['Pass']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

print("Decision Tree Predictions:", y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
rf = RandomForestClassifier(n_estimators=10, random_state=0)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Predictions:", y_pred_rf)
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
tree.plot_tree(dt, filled=True, feature_names=['Study_Hours', 'Sleep_Hours'])
plt.show()
```
